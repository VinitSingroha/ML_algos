We will try to use KNN to create a model that directly predicts a class for a new data point based off of the features.

## Standardize the Variables

Since the KNN classifier predicts the class of a given experimental observation by determining the observations closest to it, the scaling of the variables is important. All large-scale variables will have a much larger effect on the interval between observations, and therefore on the KNN classifier, than small-scale variables.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS',axis=1)) # drop the target class from the dataset
```

```python
scaled_features = scaler.transform(df.drop('TARGET CLASS',axis=1))
```

```python
## create the dataframe of scaled features
data = pd.DataFrame(scaled_features,columns=df.columns[:-1])
data.head()
```

## Split the dataset into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
## divided the dataset into 70% of train data and 30% of test data
X_train, X_test, y_train, y_test = train_test_split(data,df['TARGET CLASS'],test_size=0.30)
```

```python
X_train.shape , X_test.shape , y_train.shape , y_test.shape
```

## Implement KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3) ## choose k = 3
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test) ## prediction of test data
```

```python
pred
```

## Predictions and Evaluations of KNN model

```python
from sklearn.metrics import classification_report,confusion_matrix , plot_confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
print(confusion_matrix(y_test,pred)) ## confusion matrix
```

```python
plot_confusion_matrix(knn,X_test,y_test)
```

```python
print(classification_report(y_test,pred))
```

## Choosing a K Value by using the elbow method to pick a good K Value

```python
accuracy_rate = []
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,data,df['TARGET CLASS'],cv=10)
    accuracy_rate.append(score.mean())
```

```python
accuracy_rate
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),accuracy_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('accuracy_rate vs. K Value')
plt.xlabel('K')
plt.ylabel('accuracy_rate')
```

```python
error_rate = []
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    score=cross_val_score(knn,data,df['TARGET CLASS'],cv=10)
    error_rate.append(1-score.mean())
```

```python
error_rate
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o' ,markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='red', linestyle='dashed', marker='o',markerfacecolor='black', markersize=10)
plt.plot(range(1,40),accuracy_rate,color='blue', linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate & Accuracy Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Rate')
plt.show()
```

Here we can see that that after around K>23 the error rate just tends to hover around 0.06-0.05 Let retrain the model with that and check the classification report of model.

```python
# WITH K=23
knn = KNeighborsClassifier(n_neighbors=23)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
```

```python
print('WITH K=23')
print('\n')
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
plot_confusion_matrix(knn,X_test,y_test)
```
